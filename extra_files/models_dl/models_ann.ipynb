```python
import numpy as np 
import pandas as pd 
import tensorflow as tf
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
"""PART1 - DATA PROCESSING"""

#import dataset
df = pd.read_csv('/home/luis/workspace/covid_visualization/extra_files/dataset.csv')
X = df.iloc[:,2:-1].values  #Guarda a partir da coluna 3 (4) até a penultima
y = df.iloc[:, -1].values #Guarda apenas a ultima coluna
#One Hot Encoding
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[0])], remainder='passthrough')
X = np.array(ct.fit_transform(X)) 
#Splitting Dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
#Feature Scalling 
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print(X_train)
print(X_test)
```

```python
"""PART2 - BUILDING THE ANN"""

#Initialize ANN 
ann = tf.keras.models.Sequential()

#Adding the input layer and first 
ann.add(tf.keras.layers.Dense(units = 50, activation='relu')) # Camada Dense (Conectada a 100 %); 6 Neuronios na camada oculta e funçaõ de ativação relu

#Adding The Second Hidden Layer
ann.add(tf.keras.layers.Dense(units = 50, activation='relu')) # Camada Dense (Conectada a 100 %); 6 Neuronios na camada oculta e funçaõ de ativação relu 

#Adding The Second Hidden Layer
ann.add(tf.keras.layers.Dense(units = 50, activation='relu')) # Camada Dense (Conectada a 100 %); 6 Neuronios na camada oculta e funçaõ de ativação relu 

#Adding The Second Hidden Layer
ann.add(tf.keras.layers.Dense(units = 50, activation='relu')) # Camada Dense (Conectada a 100 %); 6 Neuronios na camada oculta e funçaõ de ativação relu 

#Adding The Output Layer 
ann.add(tf.keras.layers.Dense(units = 1, activation='sigmoid')) # 1 neurónio de saída porque é binário; funcao sigmoid para ter a probabilidade do resultado
```

```python
"""Part3 - Training The ANN """
#Compiling the Ann 
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy']) # optimezer é adam porque faz a descida do gradiente estocastico
#Training the ANN on the Training Set
history = ann.fit(X_train, y_train, batch_size = 10, epochs = 60)
```

```python
"""Predict the Test set Results"""
y_pred = ann.predict(X_test)
y_pred = y_pred > 0.5
predictons = np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1)
for i in predictons:
    print(i)
```

```python
"""Confusion Matrix"""
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
print(cm)
print(acc)
```
